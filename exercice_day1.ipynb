```python

import re

filename = "/home/pierre/import/txt/03b8f7c5006e376962b056f386b3ae93.txt"

def readtext(filename):
    with open(filename,"r") as f:
        return f.read()

def clean_text(s: str) -> list[str]:
    # tout en minuscule
    s = s.lower()
    # remplacer retour ligne / tab par espace
    s = s.replace("\n", " ").replace("\t", " ")
    # garder uniquement lettres + accents + espaces
    s = re.sub(r"[^a-zàâçéèêëîïôûùüÿñæœ\s]", " ", s)
    # réduire les espaces multiples
    s = re.sub(r"\s+", " ", s).strip()
    # découper en liste de mots
    return s.split(" ")

raw = readtext(filename)
items = clean_text(raw)
print(items)

```

```python
from pathlib import Path

dossier = Path("/home/pierre/import/txt")

allitems = []
for fichier in dossier.iterdir():
    if fichier.is_file():
        with open(fichier, "r", encoding="utf-8") as f:
            contenu = f.read()
            items = clean_text(contenu)
            
            allitems.extend(items)

print(allitems)
```

```python
def get_freq(items):
    res = {}
    for m in items:
        cpt = res.get(m,0)
        cpt += 1
        res[m] = cpt

    trie = sorted(res.items(), key=lambda x: x[1], reverse=True)
    return trie
    
```

```python
x = get_freq(allitems)
print(x)
```
